```python
novos_nomes = {
    'work_year': 'ano',
    'experience_level': 'senioridade',
    'employment_type': 'contrato',
    'job_title': 'cargo',
    'salary': 'salario',
    'salary_currency': 'moeda',
    'salary_in_usd': 'usd',
    'employee_residence': 'residencia',
    'remote_ratio': 'remoto',
    'company_location': 'empresa',
    'company_size': 'tamanho_empresa'
}
```

```python
df.rename(columns=novos_nomes, inplace=True)
```

```python
df.head()
```

```python
df["senioridade"].value_counts()
```

```python
traducao_senioridade = {
    'SE': 'Senior',
    'MI': 'Pleno',
    'EN': 'Junior',
    'EX': 'Executivo'
}
df['senioridade'] = df['senioridade'].replace(traducao_senioridade)
display(df['senioridade'].value_counts())
```

```python
df["contrato"].value_counts()
```

```python
traducao_contrato = {
    'FT': 'Integral',
    'CT': 'Contrato',
    'PT': 'Parcial',
    'FL': 'Freelancer'
}
df['contrato'] = df['contrato'].replace(traducao_contrato)
display(df['contrato'].value_counts())
```

```python
df["tamanho_empresa"].value_counts()
```

```python
traducao_tamanho_empresa = {
    'M': 'Mdio',
    'L': 'Grande',
    'S': 'Pequeno'
}
df['tamanho_empresa'] = df['tamanho_empresa'].replace(traducao_tamanho_empresa)
display(df['tamanho_empresa'].value_counts())
```

```python
traducao_remoto = {
    0: 'Presencial',
    50: 'Hbrido',
    100: 'Remoto'
}
df['remoto'] = df['remoto'].replace(traducao_remoto)
display(df['remoto'].value_counts())
```

```python
import plotly.express as px
```

```python
# Criar um grfico de mdia salarial por senioridade em barras usando o plotly

senioridade_media_salario = df_limpo.groupby('senioridade')['usd'].mean().sort_values(ascending=False).reset_index()

fig = px.bar(senioridade_media_salario,
             x='senioridade',
             y='usd',
             title='Mdia Salarial por Senioridade',
             labels={'senioridade': 'Nvel de Senioridade', 'usd': 'Mdia Salarial Anual (USD)'})

fig.show()
```

```python
# Criar um grfico de pizza sobre as categorias de trabalho

remoto_contagem = df_limpo ['remoto'].value_counts().reset_index()
remoto_contagem.columns = ['tipo_trabalho', 'quantidade']

fig = px.pie(remoto_contagem,
             names='tipo_trabalho',
             values='quantidade',
             title='Distribuio de Tipo de Trabalho')

fig.show()
```

```python
# Criar um grfico de pizza sobre as categorias de trabalho

remoto_contagem = df_limpo ['remoto'].value_counts().reset_index()
remoto_contagem.columns = ['tipo_trabalho', 'quantidade']

fig = px.pie(remoto_contagem,
             names='tipo_trabalho',
             values='quantidade',
             title='Distribuio de Tipo de Trabalho',
             hole=0.5)
fig.update_traces(textinfo='percent+label')
fig.show()
```

```python
df_ds = df_limpo[df_limpo['cargo'] == 'Data Scientist']

salario_medio_ds_por_pais = df_ds.groupby('empresa')['usd'].mean().reset_index()

salario_medio_ds_por_pais = salario_medio_ds_por_pais.sort_values('usd', ascending=False)

fig = px.bar(salario_medio_ds_por_pais,
             x='empresa',
             y='usd',
             title='Mdia Salarial para Data Scientists por Pas',
             labels={'empresa': 'Pas', 'usd': 'Mdia Salarial (USD)'})

fig.show()
```

```python
import pycountry

# Funo para converter ISO-2 para ISO-3
def iso2_to_iso3(code):
    try:
        return pycountry.countries.get(alpha_2=code).alpha_3
    except:
        return None

# Criar nova coluna com cdigo ISO-3
df_limpo['residencia_iso3'] = df_limpo['residencia'].apply(iso2_to_iso3)

# Calcular mdia salarial por pas (ISO-3)
df_ds = df_limpo[df_limpo['cargo'] == 'Data Scientist']
media_ds_pais = df_ds.groupby('residencia_iso3')['usd'].mean().reset_index()

# Gerar o mapa
fig = px.choropleth(media_ds_pais,
                    locations='residencia_iso3',
                    color='usd',
                    color_continuous_scale='rdylgn',
                    title='Salrio mdio de Cientista de Dados por pas',
                    labels={'usd': 'Salrio mdio (USD)', 'residencia_iso3': 'Pas'})

fig.show()
```
